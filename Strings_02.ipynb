```python
def convert2lowercase(my_str):
    result = ""
    for char in my_str:# 1 loop over all the chars
        if ord(char)>=ord('A') and ord(char)<=ord('Z'):# 2 find if the char is capital/upper OR not
            lowerChar = chr(ord(char)+32) # 3 "a" >> "A": ascii+ 32>> lowercase
        else:
            lowerChar= char
        result+=lowerChar
    return result
convert2lowercase(my_str="This is A stRIng")
```

```python
# question
# Take a string as input.
# Replace all the space with underscore.

my_str = "The isspace() method is a built-in string method in Python that returns True if all the characters in a string are whitespace characters"
my_str.replace(" ","_")
```

```python
def change2underscore(my_str):
    result = ""
    for i in my_str:
        if i.isspace():
            result+='_'
        else:
            result+=i
    return result

change2underscore(my_str="This is A stRIng")
```

```python
# Q2. First Occurrence

'''You are given a character string A, having length N and an integer ASCII code B.

You have to tell the leftmost occurrence of the character having ASCII code equal to B, in A or report that it does not exist.'''

def first_occurance(A,B):
  str_len = len(A)
  for i in range(str_len):
    if A[i] == chr(B):
      return i
  return -1


first_occurance(A = "aabbcc", B = 98)

#Ans = 1st b occurs at index = 2
```

```python
first_occurance(A = 'bacdaaajaaa', B = 97)

#Ans : 1st a occurs at index = 1
```
